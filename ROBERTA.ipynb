```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
```

Import Dataset

```python
df = pd.read_csv('Emotion.csv')
df
```

Check Informasi dan jumlah label dari dataset

```python
print(df.info())

print(df['label'].value_counts())
```

Tidak ada data kolom yang kosong didalam dataset tersebut. Ada 6 kelas pada kolom label

Lakukan preprocessing & Encoding untuk mengubah value label menjadi numeric

```python
label = {'sadness': 0, 'anger': 1, 'love': 2, 'surprise': 3, 'fear': 4, 'joy': 5}

X = df['text'].values
y = df['label'].values

y_encode = [label[i] for i in df['label']]
y_encode
```

Split the data into training 70%, validation 15% , and test sets 15%

```python
train_size = int(len(X) * 0.7)
val_size = int(len(X) * 0.15)
test_size = len(X) - (train_size - val_size)

X_train, y_train = X[:train_size], y_encode[:train_size]
X_val, y_val = X[train_size:train_size + val_size], y_encode[train_size + val_size:]
X_test, y_test = X[train_size + val_size:], y_encode[train_size + val_size:]
```

Instal Library yang dibutuhkan

```python
pip install --upgrade keras-nlp
```

Load RoBERTa preprocessor & Create model (RoBERTa)

```python
import keras_nlp

# print(keras_nlp.models.RobertaPreprocessor.presets.keys())
preprocessor = keras_nlp.models.RobertaPreprocessor.from_preset(
    "roberta_base_en",
    sequence_length=100
)

model = keras_nlp.models.RobertaClassifier.from_preset(
    "roberta_base_en",
    num_classes=6,
    preprocessor=preprocessor,
    activation="softmax"
)
```

Train model

```python
model.fit(x = X_train, y = y_train, validation_data = (X_val, y_val), batch_size = 32, epochs = 5)
```

Untuk membuat model RoBERTa kita akan menggunakan function RobertaPreprocessor.

Parameter yang saya masukkan adalah:
Nama model yang dipakai, num_classes yang saya set 6 karena ada 6 kelas yang akan diklasifikasi, preprocessor yang berfungsi untuk menjalankan input saat nanti dilakukan fit.

Dari hasil training didapatkan accuracy yang didapatkan baik sebesar 0.9523 untuk training, namun pada validation didapatkan accuracy yang jelek sebesar 0.2357. Pada loss didapatkan sebesar 0.0958 untuk training dan pada validation didapatkan sebesar 6.2082

Dapat disimpulkan bahwa Model mengalami overfitting karena model berusaha menyempurnakan performanya pada training (mendapatkan accuracy yg sangat baik) namun pada val performanya sangat jelek karena terlalu fokus dalam menangkap pola yang ada pada train data.

Performance Analysis (precision, recall, F1-Score and accuracy)

```python
y_pred = model.predict(X_test)
y_pred = np.argmax(y_pred, axis = 1)

print(classification_report(y_test, y_pred))
```

```python
print(precision_score(y_test, y_pred, average = 'macro'))
print(recall_score(y_test, y_pred, average = 'macro'))
print(f1_score(y_test, y_pred, average = 'macro'))
print(accuracy_score(y_test, y_pred))
```

Dapat dilihat dari hasil analysis nilai precision, accuracy, recall, F1-Score sangat tinggi. Hal itu menandakan bahwa model RoBERTa dapat melakukan klasifikasi dengan sangat baik.


Model RoBERTa mampu memprediksi nilai positif dengan sangat akurat, berdasarkan nilai precision dan recall yang tinggi. Nilai F1-Score yang tinggi menunjukkan bahwa model mampu menyeimbangkan antara akurasi dan sensitivitas. Nilai accuracy yang tinggi menunjukkan bahwa model mampu melakukan tugas klasifikasi dengan baik test data.

***What is the difference between BERT and GPT? ***


BERT berfokus dalam memahami hubungan antar kata dalam kalimat, dimana dia akan menganalisis teks secara 2 arah yaitu dengan melihat kata sebelum dan sesudah. Sehingga BERT unggul dalam pemahaman makna dan cocok dipakai untuk menjawab pertanyaan dan klasifikasi teks.


GPT berfokus pada kekreatifan dalam menghasilkan teks dengan mengandalkan kata sebelumnya untuk memprediksi kata berikutnya. GPT cocok dipakai untuk menulis cerita atau terjemahan karena pandai merangkai kalimat sendiri.
